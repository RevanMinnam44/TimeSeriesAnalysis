# Example code for interpreting the Auto-correlation function (ACF)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("test_ACFplot.xlsx")
```

```python
data.head()
```

```python
data.plot('Index', 'y(t)')
#data.plot('Month', 'Rail PM')
#data.plot('Month', 'VMT (billions)')
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(data['y(t)'], lags=10)
plt.show()
```

# ACF for time-series with trend

```python
data = pd.read_excel("test_ACFplot_WithTrend.xlsx")
data.head()
```

```python
data.plot('Index', 'y(t)')
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(data['y(t)'], lags=10)
plt.show()
```

# ACF after first differencing

```python
from statsmodels.tsa.statespace.tools import diff

diffed = diff(data.iloc[:,1])
plt.plot(diffed)
plt.show()
```

```python
plot_acf(diffed, lags=10)
plt.show()
```
